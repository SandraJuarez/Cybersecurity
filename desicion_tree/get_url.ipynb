```python
import csv

lista_urls = []
lista_source_and_dest = []

# Abrir el archivo CSV
with open('../data/captura_1_final.csv', newline='') as csvfile:
    reader = csv.reader(csvfile)
    #next(reader)  # Saltar la primera fila si contiene encabezados

    # Iterar sobre las filas del archivo CSV
    for row in reader:
        # Obtener la columna que contiene información de DNS
        dns_info = row[1]  # Ajusta el índice según la ubicación de tus datos
        
        # Dividir la línea por espacios
        partes = dns_info.split()

        # Inicializar una variable booleana para verificar si se ha encontrado un URL en la línea
        encontrado_url = False

        # Buscar "A" mayúscula o "AAA" o "AAAA" o "HTTPS" en las partes de la línea
        for i, parte in enumerate(partes[:-1]):  # No procesar el último elemento para evitar IndexError
            if parte == "A" or parte == "AAA" or parte == "AAAA" or parte == "HTTPS":
                # Verificar si hay suficientes elementos en la lista para acceder a partes[i + 1]
                if i + 1 < len(partes):
                    # Tomar la siguiente palabra como el URL si no comienza con "www."
                    if not partes[i + 1].startswith("www.") and not encontrado_url:
                        print(partes[i + 1])
                        lista_urls.append(partes[i + 1])

                        lista_source_and_dest.append(row[3] + " -> " + row[4])
                        print(row[3] + " -> " + row[4])
                        encontrado_url = True  # Marcar que se ha encontrado un URL en la línea
```

```python


#save the urls in a csv file
with open('../data/urls.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["url", "type", "Source -> Destination"])
    for i in range(len(lista_urls)):
        writer.writerow([lista_urls[i],"benign", str(lista_source_and_dest[i])])
        
```
